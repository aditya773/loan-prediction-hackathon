```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("train_ctrUa4K.csv")
```

```python
df
```

```python
df.info()
```

```python
df['Gender'].fillna('Male',inplace=True)
```

```python
df['Married'].fillna('Yes',inplace=True)
```

```python
df.Dependents.fillna('0',inplace=True)
```

```python
df['Self_Employed'].fillna('No',inplace=True)
```

```python
df['LoanAmount'].fillna(method='bfill',inplace=True)
```

```python
df['Loan_Amount_Term'].fillna(360.0,inplace=True)
```

```python
df['Credit_History'].fillna(1.0,inplace=True)
```

```python
df.info()
```

```python
df['Loan_Status'][df['Loan_Status']=='Y'] = 1
```

```python
df['Loan_Status'][df['Loan_Status']=='N'] = 0
```

```python
df
```

```python
df1=df.copy()
```

```python
df.info()
```

```python
df
```

```python
df['Loan_Status'] = df['Loan_Status'].astype('int64')
```

```python
df.info()
```

```python
df.corr()
```

```python
df
```

```python
del df['Loan_ID']
del df['Gender']
del df['Married']
del df['Education']
del df['ApplicantIncome']
```

```python
df_dep = pd.get_dummies(df['Dependents'])
df = df.join(df_dep)
del df['Dependents']
```

```python
df_sel = pd.get_dummies(df['Self_Employed'])
df = df.join(df_sel)
del df['Self_Employed']
```

```python
df_pr = pd.get_dummies(df['Property_Area'])
df = df.join(df_pr)
del df['Property_Area']
```

```python
df
```

```python
x = df.drop('Loan_Status',axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=3)
```

```python
rfc.fit(x_train,y_train)
```

```python
predictions = rfc.predict(x_test)
```

```python
y_predict = rfc.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print("training accuracy:",accuracy_score(y_train,y_predict))
print("test accuracy:",accuracy_score(y_test,predictions))
```

```python
print(confusion_matrix(predictions,y_test))
```

```python
print(classification_report(predictions,y_test))
```
